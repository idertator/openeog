# 1 - Differentiation

```python
%matplotlib inline
import matplotlib as plt
plt.rcParams['figure.figsize'] = [21, 8]
plt.rcParams['figure.dpi'] = 100
```

```python
from bsp.core.io import load_study
study = load_study("data/prueba_fixed.bsp")
```

```python
from scipy.signal import medfilt

from bsp.core import denoise, differentiate

test = study[0]
channel = denoise(test.horizontal_channel)
derivative = differentiate(channel)
filtered_derivative = medfilt(derivative, 11)

plot(derivative)
plot(filtered_derivative)
```
